```python
# Load data
import pandas as pd
df = pd.read_csv("gene_expression_project.csv")

print("Dataset loaded successfully.\n")
print("Preview of data:")
print(df.head(), "\n")

# Transform data (pivot)
pivot = df.pivot(index="Gene", columns="Condition", values="Expression")

print("Pivot table (Gene vs Condition):")
print(pivot, "\n")

# Visualization
import matplotlib.pyplot as plt

pivot.plot(kind="bar")

plt.title("Gene Expression: Normal vs Cancer")
plt.ylabel("Expression level")
plt.xticks(rotation=45)

plt.show()
print("Bar chart displayed successfully.\n")

# Compute difference
diff = pivot["cancer"] - pivot["normal"]

print("Expression difference (Cancer - Normal):")
print(diff.sort_values(ascending=False).to_string())

# Biological insight
for gene in diff.index:
    if diff[gene] > 0:
        print(f"- {gene} is upregulated in cancer")
    else:
        print(f"- {gene} is downregulated in cancer")
```
